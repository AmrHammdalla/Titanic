```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic=pd.read_csv("./Titanic.csv")
titanic.head(n=10)
```

```python
titanic.drop_duplicates(inplace=True)
titanic.info()
```

```python
titanic.isna().sum()
```

```python
titanic['Survived']=titanic['Survived']==1
titanic['Pclass'].replace(1,"first",inplace=True)
titanic['Pclass'].replace(2,"second",inplace=True)
titanic['Pclass'].replace(3,"third",inplace=True)
titanic['Cabin']=titanic['Cabin'].fillna('None')
titanic.head(30)
```

```python
# detect labels in categorical variables
for col in titanic.columns.delete([0,3,5,8,9,10]):
    print(titanic[col].value_counts())
```

####  So, what do I know about dataset?

- The *titanic* dataset has 12 columns (features) and 891 rows (observations,samples).\
  *Numerical* columns are:
    - *PassengerId* (discrete) - id for each passenger 
    - *Age* (discrete) - age for each passenger
    - *SibSp* (discrete) - Brother, Sister, Stepbrother, Stepsister of passenger Aboard Titanic
    - *Parch* (discrete) - Mother or Father of Passenger Aboard Titanic
    - *Fare* (discrete) - price

  *Categorical* columns are:
    - *Survived* (0/1) - 0->False, 1->True
    - *Pclass* (1/2/3) - 1-> highest
    - *Sex* (male/female) 
    - *Embarked* (S/C/Q) - Southampton , Cherbourg , Queenstown

- There are missing values in Age (177), Embarked (2), Cabin (687).
- Most passengers booked the 3rd class ticket, is it because of low price?.
- The passengers were mainly males.
- The average age on the ship was around 30 years old.
- Around 38% survived the disaster.
- Most Passengers Embarked from Southhampton 644 our of 891.
- Fare has min value of 0 and max value of 512 with mean 32 and std 49.
- Some tickets numbers are repeated, There could be forgery.

### *People Classification

```python
non_parents_siblings=titanic[(titanic['Parch']==0) & (titanic['SibSp']==0)]
siblings_people=titanic[(titanic['SibSp']!=0)]
family=titanic[(titanic['Parch']!=0)]
children=titanic[(titanic['Age']<16)]

non_parents_siblings_count=len(non_parents_siblings)
siblings_people_count=len(siblings_people)
family_count=len(family)
children_count=len(children)

plt.figure(figsize=(10,4))

plt.subplot(1,4,1)
ax=sns.countplot(x='Sex',data=non_parents_siblings).set_title("Non Parents Nor Siblings")


plt.subplot(1,4,2)
ax=sns.countplot(x='Sex',data=siblings_people).set_title("Siblings")


plt.subplot(1,4,3)
ax=sns.countplot(x='Sex',data=family).set_title("Family")

plt.subplot(1,4,4)
ax=sns.countplot(x='Sex',data=children).set_title("Children")


print('Solo people count= ',non_parents_siblings_count)
print('Family people count= ',siblings_people_count)
print('Parent people count= ',family_count)
print('Children count= ',children_count)
```

* - Solo people count 537 (They don't have children or any siblings aboard titanic)
* - Male count 411 , Female count 126
